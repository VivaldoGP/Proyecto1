```python
import pandas as pd
import geopandas as gpd
from osgeo import ogr
import numpy as np
```

```python
world_pop = pd.read_csv("world_pop_data/API_SP.POP.TOTL_DS2_en_csv_v2_4662410.csv")
world_geom = gpd.read_file("wb_countries_admin0_10m/WB_countries_Admin0_10m.shp")
```

```python
set_geoms = set(world_geom['ISO_A3'])
set_pop = set(world_pop['Country Code'])

a = set_geoms.intersection(set_pop)
len(a)
```

```python
world_popupaltion_filtered = world_pop.loc[world_pop['Country Code'].isin(a)]
```

```python
world_geom_filtered = world_geom.loc[world_geom['ISO_A3'].isin(a)].drop_duplicates(subset=["ISO_A3"])
```

```python
world_geom_filtered.shape
```

```python
world_popupaltion_filtered.shape
```

```python
merged = pd.merge(world_geom_filtered, world_popupaltion_filtered, how='inner', left_on="ISO_A3", right_on="Country Code")
```

```python
merged.fillna(0, inplace=True)
```

```python
merged.to_file("countries_population/countries_pop.shp")
```

```python
test = merged.iloc[:,57:119]
test
```

```python
years = np.array(test.columns)
```

```python
column_names_variation = []
series_variation = []

i = 0
for year in years:
    start = years[i]
    i += 1
    end = years[i]
    value = ((test[end] / test[start])-1) * 100
    series_variation.append(value)
    column_names_variation.append("{}-{}".format(start, end))
    #print(start, end, value)
    if i == 61:
        break
    
variation_per_year = dict(zip(column_names_variation, series_variation))
```

```python
variation_per_year
```

```python
fn = r'wb_countries_admin0_10m/WB_countries_Admin0_10m.shp'
ds = ogr.Open(fn, 0)

if ds is None:
    print(f"No se pudo abrir el archivo {0}".format(fn))
    
lyr = ds.GetLayer(0)

iso_codes = []

for feat in lyr:
    iso_code = feat.GetField('ISO_A3')
    iso_codes.append(iso_code)

set_iso_codes = set(iso_codes)

del ds
```

```python
b = set_iso_codes.intersection(set_pop)
len(b)
```
